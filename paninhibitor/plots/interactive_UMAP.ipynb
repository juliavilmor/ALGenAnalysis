```python
import pandas as pd
import numpy as np
import umap
```

```python
import umap.plot
```

```python
from MolecularAnalysis import plotlib, mollib

initial_specific = mollib.MolDB(smiDB='../round_1/Mpro_initial_specificset.smi', verbose=False)
db1 = mollib.MolDB(smiDB='../round_1/all_trial_generated_smiles_threshold.smi', verbose=False)
#db1.filter_similarity(simthreshold=1,fingerprint='Morgan4',verbose=False)
db2 = mollib.MolDB(smiDB='../round_2/all_trial_generated_smiles_threshold.smi', verbose=False)
#db2.filter_similarity(simthreshold=1,fingerprint='Morgan4',verbose=False)
db3 = mollib.MolDB(smiDB='../round_3_reducedta/all_trial_generated_smiles_threshold.smi', verbose=False)
#db3.filter_similarity(simthreshold=1,fingerprint='Morgan4',verbose=False)
db4 = mollib.MolDB(smiDB='../round_4/all_trial_generated_smiles_threshold.smi', verbose = False)
#db4.filter_similarity(simthreshold=1,fingerprint='Morgan4',verbose=False)

fingers = db1.get_fingerprints(alg='RDKIT', random_max=None,verbose=False)
print(len(fingers))
print(len(list(db1.smiles)))
smiles = list(initial_specific.smiles) + list(db1.smiles) + list(db2.smiles) + list(db3.smiles) + list(db4.smiles)
print(len(smiles))
```

```python
# Plot reducer --> random_max to max to get all SMILES (M)
X, Y, S, A, M = plotlib._prepare_reducer(dbs=[db4, db3, db2, db1, initial_specific], names=['outer4', 'outer3', 'outer2', 'outer1', 'specific'], random_max=100000, delimiter=None, fpsalg='Morgan4', sizes=[0.5, 0.5, 0.5, 0.5, 0.5], alphas=[0.9, 0.9, 0,9, 0.9, 0.9])

mapper = umap.UMAP().fit(X)
```

```python
# get the IDs for each molecule
all_mols = mollib.MolDB(sdfDB='/home/cactus/julia/results/Mpro_target_sel/separate_peptides/Mpro_nonpeptides.sdf', verbose=False)
map = {}
for k in all_mols.dicDB.keys():
    ID = all_mols.dicDB[k][1]
    smile = all_mols.dicDB[k][0]
    map[ID] = smile


# Hover data is the information that will appear interactively in the UMAP
hover_data = pd.DataFrame({'index': np.arange(X.shape[0]),
                          'label': Y,
                          'smile': M,
                          })
hover_data['ID'] = hover_data['smile'].map(dict)


print(hover_data)
```

```python
from bokeh.plotting import *
umap.plot.output_notebook()
p = umap.plot.interactive(mapper, labels=Y, hover_data=hover_data, point_size=2, color_key_cmap='YlGnBu')
show(p)
```
